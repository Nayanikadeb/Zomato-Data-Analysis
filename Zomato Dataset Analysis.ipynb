```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('C:\\Users\\LENOVO\\OneDrive\\Documents\\Data analysis practice dataset\\zomato.csv')
df.head()
```

```python
df.columns
```

```python
df=df.drop(['url','address','phone','menu_item','dish_liked','reviews_list'],axis=1)
df.head()
```

```python
df.info()
```

### Dropping duplicates

```python
df.drop_duplicates(inplace=True)
df.shape
```

### Cleaning rate column 

```python
df.rate.unique()
```

```python
def cleanrate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
df['rate']=df['rate'].apply(cleanrate)
df['rate'].head()
```

### Filling null values in rate column with the mean 

```python
df['rate'].fillna(df['rate'].mean(), inplace=True)
df['rate'].isnull().sum()
```

```python
df.info()
```

### Dropping the remaining null values( because there aren't too many)

```python
df.dropna(inplace=True)
df.head()
```

```python
df.rename(columns = {'approx_cost(for two people)':'Cost2plates', 'listed_in(type)':'Type'}, inplace = True)
df.head()
```

Since location and listed_in(city) both contains same location, we will drop one column

```python
df=df.drop(['listed_in(city)'],axis=1)
df.head()
```

### Cleaning Cost2plates column

```python
def removecomma(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)
df['Cost2plates']=df['Cost2plates'].apply(removecomma)
df['Cost2plates'].unique()    
```

```python
df['rest_type'].value_counts()
rest_types = df['rest_type'].value_counts(ascending  = False)
rest_types_lessthan1000 = rest_types[rest_types<1000]
```

### Cleaning rest_type column 

Some rest_type has too less number of restaurants. So we will club them into a category called 'Others'

```python
def clean_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'Others'
    else:
        return value
df['rest_type']=df['rest_type'].apply(clean_rest_type)
df['rest_type'].value_counts()
```

### Cleaning location same way as rest_type column

```python
df['location'].value_counts()
locations = df['location'].value_counts(ascending  = False)
locations_lessthan300 = locations[locations<300]
```

```python
def clean_location(value):
    if(value in locations_lessthan300):
        return 'Others'
    else:
        return value
df['location']=df['location'].apply(clean_location)
df['location'].value_counts()
```

### Cleaning cuisine same way as rest_type column

```python
df['cuisines'].value_counts()
cuisine_total = df['cuisines'].value_counts(ascending  = False)
cuisines_lessthan100 = cuisine_total[cuisine_total<100]
```

```python
def clean_cuisine(value):
    if(value in cuisines_lessthan100):
        return 'Others'
    else:
        return value
df['cuisines']=df['cuisines'].apply(clean_cuisine)
df['cuisines'].value_counts()
```
